```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Raw dataset
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nahn, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
#Task 1: Clean the Data
# Impute missing values
df['age'].fillna(df['age'].median(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)

# Remove duplicates
df = df.drop_duplicates()

print("Cleaned Data:\n", df)
```

```python
df
```

```python
df.info()
```

```python
# Task 2: Encode Categorical Columns
# One-Hot Encoding for city
df = pd.get_dummies(df, columns=['city'], prefix='city')

# Label Encoding for gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print("\nEncoded Data:\n", df)
```

```python
df.columns
```

```python
#Task 3: Scale Numerical Features
# Select numerical columns
num_cols = ['age', 'annual_income']

# Min-Max Scaling
minmax_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = minmax_scaler.fit_transform(df[num_cols])
print("\nMin-Max Scaled Data:\n", df_minmax)

# Standardisation
std_scaler = StandardScaler()
df_std = df.copy()
df_std[num_cols] = std_scaler.fit_transform(df[num_cols])
print("\nStandardised Data:\n", df_std)
```
